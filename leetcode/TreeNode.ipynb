```python
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def __init__(self):
        self.record = []
        
    def backtrack(self, node):
        if not node:
            return
        self.backtrack(node.left)
        self.record.append(node.val)
        self.backtrack(node.right)
        
    def inorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        self.backtrack(root)
        return self.record

s = Solution()
s.inorderTraversal(root)
```

```python
#include <vector>
#include <iostream>

struct TreeNode {
    int val;
    TreeNode *left;
    TreeNode *right;
    TreeNode() : val(0), left(nullptr), right(nullptr) {}
    TreeNode(int x) : val(x), left(nullptr), right(nullptr) {}
    TreeNode(int x, TreeNode *left, TreeNode *right) : val(x), left(left), right(right) {}
};

class Solution {
public:
    std::vector<int> record;
    void backtrack(TreeNode* node) {
        if (!node) {
            return;
        }
        backtrack(node->left);
        record.emplace_back(node->val);
        backtrack(node->right);
    }
    std::vector<int> inorderTraversal(TreeNode* root) {
        backtrack(root);
        return record;
    }
};

int main()
{
    TreeNode* root = new TreeNode{ 5 };
    root->left = new TreeNode{ 4 };
    root->right = new TreeNode{ 8 };
    root->left->left = new TreeNode{ 11 };
    root->left->left->left = new TreeNode{ 7 };
    root->left->left->right = new TreeNode{ 2 };
    root->right->left = new TreeNode{ 13 };
    root->right->right = new TreeNode{ 4 };
    root->right->right->left = new TreeNode{ 5 };
    root->right->right->right = new TreeNode{ 1 };
    
    Solution s;
    std::vector<int> output = s.inorderTraversal(root);

    for (int ii = 0; ii < output.size(); ii++) {
        std::cout << output[ii] << ", ";
    }
}
```

```python
class Solution(object):
    def levelOrder(self, root):
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        output = []
        record = []
        if root:
            record.append(root)
        while record:
            val_in_cur_layer = []
            record_new = []
            for node in record:
                val_in_cur_layer.append(node.val)
                if node.left:
                    record_new.append(node.left)
                if node.right:
                    record_new.append(node.right)
            output.append(val_in_cur_layer)
            record = record_new
        return output

s = Solution()
s.levelOrder(root)
```

```python
#include <vector>
#include <iostream>

struct TreeNode {
    int val;
    TreeNode* left;
    TreeNode* right;
    TreeNode() : val(0), left(nullptr), right(nullptr) {}
    TreeNode(int x) : val(x), left(nullptr), right(nullptr) {}
    TreeNode(int x, TreeNode* left, TreeNode* right) : val(x), left(left), right(right) {}
};

class Solution {
public:
    std::vector<std::vector<int>> levelOrder(TreeNode* root) {
        std::vector<std::vector<int>> output;
        std::vector<TreeNode*> record;
        if (root) {
            record.emplace_back(root);
        }
        while (record.size() > 0) {
            std::vector<TreeNode*> record_new;
            std::vector<int> val_in_cur_layer;
            for (auto node : record) {
                val_in_cur_layer.emplace_back(node->val);
                if (node->left) {
                    record_new.emplace_back(node->left);
                }
                if (node->right) {
                    record_new.emplace_back(node->right);
                }
            }
            output.emplace_back(val_in_cur_layer);
            record.swap(record_new);
        }
        return output;
    }
};

int main()
{
    TreeNode* root = new TreeNode{ 5 };
    root->left = new TreeNode{ 4 };
    root->right = new TreeNode{ 8 };
    root->left->left = new TreeNode{ 11 };
    root->left->left->left = new TreeNode{ 7 };
    root->left->left->right = new TreeNode{ 2 };
    root->right->left = new TreeNode{ 13 };
    root->right->right = new TreeNode{ 4 };
    root->right->right->left = new TreeNode{ 5 };
    root->right->right->right = new TreeNode{ 1 };

    Solution s;
    std::vector<std::vector<int>> output = s.levelOrder(root);

    for (int ii = 0; ii < output.size(); ii++) {
        for (int jj = 0; jj < output[ii].size(); jj++) {
            std::cout << output[ii][jj] << ", ";
        }
        std::cout << std::endl;
    }
}
```

```python
record = [root]
label = True
while len(record) > 0:
    temp_val = []
    temp_node = []
    for _ in range(len(record)):
        cur_node = record.pop(0)
        if cur_node:
            temp_val.append(cur_node.val)
            record.append(cur_node.left)
            record.append(cur_node.right)
        else:
            temp_val.append(None)
    if len(temp_val) > 1:
        index_left = 0
        index_right = len(temp_val)-1
        while index_left < index_right:
            if temp_val[index_left] == temp_val[index_right]:
                index_left += 1
                index_right -= 1
            else:
                label = False
                break
    if label == False:
        break        
print(label)
```

```python
def dfs(left,right):
    # 递归的终止条件是两个节点都为空
    # 或者两个节点中有一个为空
    # 或者两个节点的值不相等
    if not (left or right):
        return True
    if not (left and right):
        return False
    if left.val!=right.val:
        return False
    return dfs(left.left,right.right) and dfs(left.right,right.left)

def isSymmetric(root):
    if not root:
        return True
    # 用递归函数，比较左节点，右节点
    return dfs(root.left,root.right)

isSymmetric(root)
```

## [二叉树的最大深度](https://leetcode.cn/problems/maximum-depth-of-binary-tree/)
$\iff$ <span id='LCR.175'></span>[LCR 175. 计算二叉树的深度](https://leetcode.cn/problems/er-cha-shu-de-shen-du-lcof/)  
给定一个二叉树，找出其最大深度。

二叉树的深度为根节点到最远叶子节点的最长路径上的节点数。

说明: 叶子节点是指没有子节点的节点。

```python
def maxDepth(root):
    if not root:
        return 0
    else:
        return max(maxDepth(root.left), maxDepth(root.right)) + 1
maxDepth(root)
```

### [110. 平衡二叉树](https://leetcode.cn/problems/balanced-binary-tree/)
$\iff$ [LCR 176. 判断是否为平衡二叉树](https://leetcode.cn/problems/ping-heng-er-cha-shu-lcof/)  
给定一个二叉树，判断它是否是高度平衡的二叉树。

本题中，一棵高度平衡二叉树定义为: **一个二叉树每个节点 的左右两个子树的高度差的绝对值不超过 1**。


```python
def maxDepth(root):
    if not root:
        return 0
    else:
        return max(maxDepth(root.left), maxDepth(root.right)) + 1
def isBalanced(root):
    if not root:
        return True
    else:        
        return abs(maxDepth(root.left) - maxDepth(root.right)) < 2 and isBalanced(root.left) and isBalanced(root.right)

isBalanced(root)
```

```python
/**
 * Definition for a binary tree node.
 * struct TreeNode {
 *     int val;
 *     TreeNode *left;
 *     TreeNode *right;
 *     TreeNode() : val(0), left(nullptr), right(nullptr) {}
 *     TreeNode(int x) : val(x), left(nullptr), right(nullptr) {}
 *     TreeNode(int x, TreeNode *left, TreeNode *right) : val(x), left(left), right(right) {}
 * };
 */
class Solution {
public:
    int maxDeep(TreeNode* root) {
        if (!root) {
            return 0;
        } else {
            return max(maxDeep(root->left), maxDeep(root->right)) + 1;
        }
    }
    bool isBalanced(TreeNode* root) {
        if (!root) {
            return true;
        } else {
            return abs(maxDeep(root->left) - maxDeep(root->right)) < 2 && isBalanced(root->left) && isBalanced(root->right);
        }
    }
};
```

### [437. 路径总和 III](https://leetcode-cn.com/problems/path-sum-iii/)  
$\iff$ <span id='Offer.II050'></span>[剑指 Offer II 050. 向下的路径节点之和](https://leetcode-cn.com/problems/6eUYwP/)  

给定一个二叉树的根节点$root$，和一个整数$targetSum$，求该二叉树里节点值之和等于$targetSum$的**路径**的数目。

**路径**不需要从根节点开始，也不需要在叶子节点结束，但是路径方向必须是向下的（只能从父节点到子节点）。


```python
targetSum = 9

def dfs(root, visited):
    if not root:
        return 0
    else:
        count = 0
        temp = visited + [root.val]
        Sum = 0
        for ii in range(len(temp)-1,-1,-1):
            Sum += temp[ii]
            if Sum == targetSum:
                count += 1
        if root.left != None:
            count += dfs(root.left, temp)
        if root.right != None:
            count += dfs(root.right, temp)
        return count

count = dfs(root, [])
print(count)
```

```python
// cpp code
/*
* Definition for a binary tree node.
* struct TreeNode {
*     int val;
*     TreeNode *left;
*     TreeNode *right;
*     TreeNode() : val(0), left(nullptr), right(nullptr) {}
*     TreeNode(int x) : val(x), left(nullptr), right(nullptr) {}
*     TreeNode(int x, TreeNode *left, TreeNode *right) : val(x), left(left), right(right) {}
* };
*/

class Solution {
public:
    int count = 0;
    
    void dfs(TreeNode* root, vector<int>& record, int targetSum){
        if (root == nullptr) {
            return;
        }
        else {
            vector<int> temp;
            temp = record;
            temp.emplace_back(root->val);
            long long Sum = 0;
            for (int ii=temp.size()-1;ii>-1;ii--) {
                Sum += temp[ii];
                if (Sum == targetSum) {
                    count += 1;
                }
            }
            if (root->left != nullptr) {
                dfs(root->left, temp, targetSum);
            }
            if (root->right != nullptr) {
                dfs(root->right, temp, targetSum);
            }
        }
    }
    int pathSum(TreeNode* root, int targetSum) {
        vector<int> record;
        dfs(root, record, targetSum);
        return count;
    }
};
```

## [563. 二叉树的坡度](https://leetcode.cn/problems/binary-tree-tilt/description/)
给定一个二叉树，计算 整个树 的坡度 。

一个树的 节点的坡度 定义即为，该节点左子树的节点之和和右子树节点之和的 差的绝对值 。如果没有左子树的话，左子树的节点之和为 0 ；没有右子树的话也是一样。空结点的坡度是 0 。

整个树 的坡度就是其所有节点的坡度之和。

  
输入：root = [4,2,9,3,5,null,7]  
输出：15  
解释：  
节点 3 的坡度：|0-0| = 0（没有子节点）  
节点 5 的坡度：|0-0| = 0（没有子节点）  
节点 7 的坡度：|0-0| = 0（没有子节点）  
节点 2 的坡度：|3-5| = 2（左子树就是左子节点，所以和是 3 ；右子树就是右子节点，所以和是 5 ）  
节点 9 的坡度：|0-7| = 7（没有左子树，所以和是 0 ；右子树正好是右子节点，所以和是 7 ）  
节点 4 的坡度：|(3+5+2)-(9+7)| = |10-16| = 6（左子树值为 3、5 和 2 ，和是 10 ；右子树值为 9 和 7 ，和是 16 ）  
坡度总和：0 + 0 + 0 + 2 + 7 + 6 = 15


输入：root = [21,7,14,1,1,2,2,3,3]  
输出：9  

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def findTilt(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        def Sum(root):
            if root == None:
                return 0
            else:
                return Sum(root.left) + Sum(root.right) + root.val
        if root == None:
            return 0
        else:
            return self.findTilt(root.left) + self.findTilt(root.right) + abs(Sum(root.left)-Sum(root.right))
s = Solution()
s.findTilt(root)
```
