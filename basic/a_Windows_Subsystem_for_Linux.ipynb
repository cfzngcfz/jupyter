### 修改文件/文件夹的权限
refer to: https://www.cnblogs.com/sxdcgaq8080/p/7498906.html  

字符|数字|描述
:-:|:-:|:-
-rw------- | (600) | 只有所有者才有读和写的权限  
-rw-r--r-- | (644) | 只有所有者才有读和写的权限，组群和其他人只有读的权限  
-rwx------ | (700) | 只有所有者才有读，写，执行的权限  
-rwxr-xr-x | (755) | 只有所有者才有读，写，执行的权限，组群和其他人只有读和执行的权限  
-rwx--x--x | (711) | 只有所有者才有读，写，执行的权限，组群和其他人只有执行的权限  
-rw-rw-rw- | (666) | 每个人都有读写的权限  
-rwxrwxrwx | (777) | 每个人都有读写和执行的权限  

```python
sudo chmod 700 <file_name>  
sudo chmod -R 700 <folder_name>
```

不安装libgtk2.0-dev，则会报如下错误:  
terminate called after throwing an instance of 'cv::Exception'

## 合并若干commit
假如希望将最近的3个commit合并成1个commit, 合并前的git log如下:

```python
commit 22ed528c4654d9e4487bbdc2a4291dda9497ce5a (HEAD -> user/cfz/test)
Date:   Thu Dec 14 17:22:55 2023 +0800

    commit 3

commit 680a8aa6582aa228a509aecaccd8557dfd202395
Date:   Thu Dec 14 17:22:43 2023 +0800

    commit 2

commit b94fe118157b1d6902f4d74f91356c1829c978ae
Date:   Thu Dec 14 17:22:03 2023 +0800

    commit 1

commit 838030dae0d39ead4407115cc572d187c4745764 (origin/user/cfz/test)
Date:   Tue Nov 7 19:10:40 2023 +0800

    in process
```

### 合并方法1

```python
# 查看最近4个commit
git log -4
# 从版本库回退内容到暂存区, 不改动工作区的内容
git reset --soft 838030dae0d39ead4407115cc572d187c4745764 # 别人改的commitID
# add已跟踪的所有文件
git add -u
# 提交合并后的commit
git commit -m "multiple commits"
# 强制推送到远端仓库
git push -f
```

## 报错&解决
### ssh: connect to host github.com port 22: Connection timed out
refer to: https://www.jianshu.com/p/c3aac5024877
1. 测试可用性



```python
ssh -T -p 443 git@ssh.github.com
```

如果返回`Hi cfzngcfz! You've successfully authenticated, but GitHub does not provide shell access.`, 则可用

2. 编辑 ~/.ssh/config 文件，如果config文件不存在, 直接`vim ~/.ssh/config`加入以下内容

```python
Host github.com
Hostname ssh.github.com
Port 443
```

3. 验证

```python
ssh -T git@github.com
```

如果返回`Hi cfzngcfz! You've successfully authenticated, but GitHub does not provide shell access.`, 则成功
