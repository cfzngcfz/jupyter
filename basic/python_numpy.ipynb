# vstack 垂直拼接

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
np.vstack((a,b))
```

```python
a = np.array([[1], [2], [3]])
b = np.array([[4], [5], [6]])
np.vstack((a,b))
```

# hstack 水平拼接

```python
a = np.array((1,2,3))
b = np.array((4,5,6))
np.hstack((a,b))
```

```python
a = np.array([[1],[2],[3]])
b = np.array([[4],[5],[6]])
np.hstack((a,b))
```

```python
a = np.array([[] for _ in range(3)])
b = np.array([4,5,6])
print(b, b.shape)
print(b.T, b.T.shape)
print(b[:,None], b[:,None].shape)

a = np.hstack( (a,b[:,None]) )
#a = np.hstack( a,b[:,None] ) # 报错

print(a)
```

# 数列
## 等差数列
### np.arange(start=0, stop, step=1, dtype)
参数|描述
:-|:-
start|起始值，默认为0
stop|终止值(不包含)
step|步长，默认为1
dtype|返回的数据类型

```python
np.arange(0,10,1)
```

### np.linspace(start, stop, num, endpoint=True, retstep=False, dtype)


参数|描述
:-|:-
start|起始值
stop|终止值
num|等步长的样本数量，默认为50
endpoint|如果值为True，则数列包含stop，否则不包含，默认为True
retstep|如果值为True，则结果输出步长，否则不显示
dtype|返回的数据类型

```python
np.linspace(0,1,11)
```

```python
np.linspace(0,1,11,endpoint=False)
```

## 等比数列
### np.logspace(start, stop, num=50, endpoint=True, base=10.0, dtype=None)

参数|描述
:-|:-
start|起始值$\to base^{start}$
stop|终止值$\to base^{stop}$
num|等步长的样本数量，默认为50
endpoint|如果值为True，数列包含stop，否则不包含，默认为True
base|对数log的底数
dtype|返回的数据类型

```python
np.logspace(0., 9., num=10, endpoint=True, base=2.0)
```
