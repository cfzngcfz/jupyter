# 时间差

```python
from datetime import timedelta
delta = timedelta(weeks=2, days=50, hours=8, minutes=5, seconds=27, microseconds=10, milliseconds=29000)
print(delta)
print(delta.total_seconds()) #返回时间间隔包含了多少秒
print('timedelta下限:', timedelta.min)
print('timedelta上限:', timedelta.max)
print('timedelta单位:', timedelta.resolution)
```

# date

```python
from datetime import date
date01 = date(2019, 12, 31)
date01_str = '2019-12-31'

print(date01,type(date01))
print(date01_str, type(date01_str))
print('year:',date01.year)
print('month:',date01.month)
print('day:',date01.day)
print('星期',date01.weekday()+1)
print('星期',date01.isoweekday()) #星期二
print('英文缩写表达:',date01.ctime(), type(date01.ctime()))
print('date所有属性(本地):',date01.timetuple())
print('year',date01.timetuple()[0])
print('month',date01.timetuple()[1])
print('day',date01.timetuple()[2])
print('hour',date01.timetuple()[3])
print('minute',date01.timetuple()[4])
print('second',date01.timetuple()[5])
print('星期',date01.timetuple()[6]+1)
print('第',date01.timetuple()[7],'天')

print(date.today()) #当日

print('date下限:', date.min)
print('date上限:', date.max)
print('date单位:', date.resolution)
```

# time

```python
from datetime import time,timezone
date02 = time(23, 59, 58, tzinfo=timezone(timedelta(hours=8)))
date02_str = '23:59:58+08:00'

print(date02,type(date02))
print('hour:',date02.hour)
print('minute:',date02.minute)
print('second:',date02.second)
print('microsecond:',date02.microsecond)
print('tzinfo:',date02.tzinfo) #时区

print('time下限:', time.min)
print('time上限:', time.max)
print('time单位:', time.resolution)
```

# datetime

```python
from datetime import datetime,timezone
date03 = datetime(2019, 12, 31, 23, 59, 58, tzinfo=timezone(timedelta(hours=8)))
date03_str = '2019-12-31 23:59:58+08:00'

print(date03, type(date03))
print(date03_str, type(date03_str))
print('year:',date03.year)
print('month:',date03.month)
print('day:',date03.day)
print('hour:',date03.hour)
print('minute:',date03.minute)
print('second:',date03.second)
print('microsecond:',date03.microsecond)
print('tzinfo:',date03.tzinfo) #时区
print('星期',date03.weekday()+1)
print('星期',date03.isoweekday()) #星期二
print('英文缩写表达:',date03.ctime(), type(date03.ctime()))
print('datetime所有属性(本地):',date03.timetuple())
print('datetime所有属性(UTC):',date03.utctimetuple())
print('year:',date03.timetuple()[0])
print('month:',date03.timetuple()[1])
print('day:',date03.timetuple()[2])
print('hour:',date03.timetuple()[3])
print('minute:',date03.timetuple()[4])
print('second:',date03.timetuple()[5])
print('星期',date03.timetuple()[6]+1)
print('第',date03.timetuple()[7],'天')

print(date.today()) # 目前本地时间
print(datetime.today()) # 目前本地时间
print(datetime.now()) # 目前本地时间
print(datetime.now(timezone.utc)) # 目前UTC标准时间
print(datetime.utcnow()) # 目前UTC标准时间

print('datetime下限:', datetime.min)
print('datetime上限:', datetime.max)
print('datetime单位:', datetime.resolution)
```

# 时间序列

```python
# 5.1.date/datetime对象 --> int，格列高利历序号，其中公元1年1月1日的序号为1
datetime_index01 = date.toordinal(date01) 
print('index of day:', datetime_index01, type(datetime_index01))
datetime_index02 = date.toordinal(date03) 
print('index of day:', datetime_index02, type(datetime_index02))

# 5.2.int --> date/datetime对象，公元1年1月1日后的第几天
print(date.fromordinal(datetime_index01), type(date.fromordinal(datetime_index01)))
print(datetime.fromordinal(datetime_index02), type(datetime.fromordinal(datetime_index02))) #hour, minute, second 和 microsecond 值均为 0

# 5.3.datetime对象 --> POSIX时间戳(float); 仅针对datetime对象有效
print('index of second:', datetime.timestamp(date03),
      date03.timestamp() == datetime.timestamp(date03))
datetime_index03 = date03.timestamp()

# 5.4.POSIX时间戳(float) --> date/datetime对象
print(date.fromtimestamp(datetime_index03))
print(datetime.fromtimestamp(datetime_index03))
print(datetime.utcfromtimestamp(datetime_index03))
```

# 格式转换

```python
# from datetime import datetime, date, time, timezone
# 6.1.date/time/datetime对象 --> 字符串
# 字符串默认格式: YYYY-MM-DD[THH[:MM[:SS[.fff[fff]]]][+HH:MM]]
print(date.isoformat(date01), type(date.isoformat(date01)),
      date.isoformat(date01)==date01.isoformat(),
      date.isoformat(date01)==date01.__str__(),
      date01.__str__()==date.__str__(date01) )
print(time.isoformat(date02), type(time.isoformat(date02)),
      time.isoformat(date02)==date02.isoformat(),
      date02.isoformat()==date02.isoformat(timespec='auto'),
      date02.isoformat(timespec='auto')==date02.isoformat(timespec='seconds'),
      time.isoformat(date02)==date02.__str__(),
      date02.__str__()==time.__str__(date02) )
print(date02.isoformat(timespec='minutes'))
print(date02.isoformat(timespec='microseconds'))
print(datetime.isoformat(date03), type(datetime.isoformat(date03)),
      datetime.isoformat(date03)==date03.isoformat(),
      date03.isoformat()==date03.isoformat(sep='T', timespec='auto') )
print(date03.isoformat(sep=' ', timespec='auto'), type(date03.isoformat(sep=' ', timespec='auto')),
      date03.isoformat(sep=' ', timespec='auto')==date03.isoformat(sep=' ', timespec='seconds'),
      date03.isoformat(sep=' ', timespec='seconds')==date03.__str__(),
      date03.__str__()==datetime.__str__(date03))
print(date03.isoformat(sep=' ', timespec='minutes'))
print(date03.isoformat(sep=' ', timespec='microseconds'))
# 'sep'表示day与hour之间的分隔符; 'timespec'表示最小时间价格，默认为seconds

# 6.2.字符串 --> date/time/datetime对象
# 字符串格式: YYYY-MM-DD[*HH[:MM[:SS[.fff[fff]]]][+HH:MM[:SS[.ffffff]]]], 其中*可匹配任意单个字符
print(date.fromisoformat(date01_str), type(date.fromisoformat(date01_str)))
print(time.fromisoformat(date02_str), type(time.fromisoformat(date02_str)))
print(datetime.fromisoformat(date03_str), type(datetime.fromisoformat(date03_str)))
#......................................#
print(datetime.fromisoformat('2011-11-04'))
print(datetime.fromisoformat('2011-11-04T00:05:23'))
print(datetime.fromisoformat('2011-11-04 00:05:23.283975'))
print(datetime.fromisoformat('2011-11-04 00:05:23.283+00:00'))
print(datetime.fromisoformat('2011-11-04T00:05:23+04:00'))

# 6.3.date/time/datetime对象 --> 特定格式字符串
print(date01.strftime('%A, %d. %B %Y'), type(date01.strftime('%A, %d. %B %Y')),
      date01.strftime('%A, %d. %B %Y')==date.strftime(date01, '%A, %d. %B %Y'),
      date01.strftime('%A, %d. %B %Y')==date01.__format__('%A, %d. %B %Y'),
      date01.__format__('%A, %d. %B %Y')==date.__format__(date01, '%A, %d. %B %Y') )
print(date02.strftime('%I:%M%p'), type(date02.strftime('%I:%M%p')),
      date02.strftime('%I:%M%p')==time.strftime(date02, '%I:%M%p'),
      date02.strftime('%I:%M%p')==date02.__format__('%I:%M%p'),
      date02.__format__('%I:%M%p')==time.__format__(date02, '%I:%M%p') )
print(date03.strftime('%A, %d. %B %Y %I:%M%p'), type(date03.strftime('%A, %d. %B %Y %I:%M%p')),
      date03.strftime('%A, %d. %B %Y %I:%M%p')==datetime.strftime(date03, '%A, %d. %B %Y %I:%M%p'),
      date03.strftime('%A, %d. %B %Y %I:%M%p')==date03.__format__('%A, %d. %B %Y %I:%M%p'),
      date03.__format__('%A, %d. %B %Y %I:%M%p')==datetime.__format__(date03, '%A, %d. %B %Y %I:%M%p') )
#......................................#
date07 = datetime(2018, 1, 1, 23, 2, 3, 83, tzinfo=timezone(timedelta(hours=8))) #周1
print(date07.strftime('%A %B %d, %Y | %I:%M:%S.%f %p (%Z)'), type(date07.strftime('%A %B %d, %Y | %I:%M:%S.%f %p (%Z)')) )
print(date07.strftime('%a. %b. %d, %y %% %H:%M:%S (%z)'), type(date07.strftime('%a. %b. %d, %y %% %H:%M:%S (%z)')) )
#......................................#
print('year:', date07.strftime('%Y')) # Y 年份(四位十进制)
print('year:', date07.strftime('%y')) # y 年份(两位十进制)
print('year:', date07.strftime('%G')) # G 年份(四位十进制)
#......................................#
print('month:', date07.strftime('%m')) # m 月份(两位十进制)
print('month:', date07.strftime('%B')) # B 月份的英文全称
print('month:', date07.strftime('%b')) # b 月份的英文简写
#......................................#
print('当月第', date07.strftime('%d'), '天') # d 当月第几天(两位十进制)
print('当年第', date07.strftime('%j'), '天') # j 当年第几天(三位十进制)
print('当年第', date07.strftime('%W'), '周') # W 当年的第几周(两位十进制，星期一作为每周的第一天，如果第一个星期1之前存在其他日期，则这些日期被视为在第0周)
print('当年第', date07.strftime('%U'), '周') # U 当年的第几周(两位十进制，星期日作为每周的第一天，如果第一个星期7之前存在其他日期，则这些日期被视为在第0周)
print('当年第', date07.strftime('%V'), '周') # W 当年的第几周(两位十进制，星期一作为每周的第一天，包含1月4日的星期为第1周)
#......................................#
print('星期', date07.strftime('%w')) # w 星期几(一位十进制, 0->星期日)
print('星期', date07.strftime('%u')) # u 星期几(一位十进制, 7->星期日)
print('星期', date07.strftime('%A')) # A 星期几的英文全称
print('星期', date07.strftime('%a')) # a 星期几的英文简写
#......................................#
print('hour:', date07.strftime('%H')) # H 24小时制的小时(两位十进制)
print('hour:', date07.strftime('%I')) # I 12小时制的小时(两位十进制)
print('hour:', date07.strftime('%p')) # p AM(晚上12点到中午12点) or PM(中午12点到晚上12点)
#......................................#
print('minute:', date07.strftime('%M')) # M 分钟(两位十进制)
print('second:', date07.strftime('%S')) # S 秒(两位十进制)
print('microsecond:', date07.strftime('%f')) # f 毫秒(六位十进制)
#......................................#
print('local datetime:', date07.strftime('%c')) # c 本地date+time
print('local date:', date07.strftime('%x')) # x 本地date
print('local time:', date07.strftime('%X')) # X 本地time
print('时区:', date07.strftime('%z')) # z UTC偏移量(±HHMM[SS[.ffffff]])
print('时区:', date07.strftime('%Z')) # Z 时区名称
#......................................#
print(date07.strftime('%%')) # 字符'%'

# 6.4.特定格式字符串 --> datetime对象
date07_str = '2018-01-01 23:02:03.000083+08:00'
print(datetime.strptime(date07_str, '%Y-%m-%d %H:%M:%S.%f%z'), type(datetime.strptime(date07_str, '%Y-%m-%d %H:%M:%S.%f%z')) )
```

# 修改和组合

```python
# replace
print(date01, ' --> ', date01.replace(month=1, day=2), type(date01.replace(month=1, day=2)))
print(date02, ' --> ', date02.replace(hour=3, minute=4), type(date02.replace(hour=3, minute=4))) 
print(date03, ' --> ', date03.replace(day=2).replace(hour=3), type(date03.replace(day=2).replace(hour=3))) 

# date + time --> datetime
date04 = date(2011,11,4)
date05 = time(0,5,23,283,tzinfo=timezone.utc)
date06 = datetime(2011,11,4,0,5,23,283,tzinfo=timezone.utc)
print(date06,
      datetime.combine(date04, date05) == date06)
# datetime --> date
print(date06.date(),
      date06.date() == date04)
# datetime --> time
print(date06.time(), #不包含时区信息
      date06.time() == date05)
print(date06.timetz(), #包含时区信息
      date06.timetz() == date05)
```

# 时区

```python
from datetime import datetime, time, timezone
# 8.1.时区表示法
print('伦敦(中时区):', timezone.utc,
      timezone.utc== timezone(timedelta(0)) )
print('北京(东八区):', timezone(timedelta(hours=8)) )

# 8.2.时区信息(for time/datetime)
print('时区信息 --> 时间:', date02.utcoffset(),
      date02.utcoffset()==time.utcoffset(date02),
      '; 名称', date02.tzname(),
      date02.tzname()==time.tzname(date02) )
print('时区信息 --> 时间:', date03.utcoffset(),
      date03.utcoffset()==datetime.utcoffset(date03),
      '; 名称', date03.tzname(),
      date03.tzname()==datetime.tzname(date03) )

# 8.3.时区转换(only for datetime)
date_temp = datetime(2019, 12, 31, 23, 59, 58, tzinfo=timezone(timedelta(hours=8)) )
print('date 1:', date_temp)
date_temp02 = date_temp.astimezone(tz=timezone.utc)
print('date 2:', date_temp02)
print('date 1 = date 2?', date_temp==date_temp02)
```
